```python
import numpy as np
a= input("Enter marks: ").split()
marks = np.array(a, dtype=int)
failCount = 0
passCount = 0
for i in marks:
    if(i>=50):
        passCount+=1
    else:
        failCount+=1
print("Number of students passed: ",passCount)
print("Number of students failed: ",failCount)

```
